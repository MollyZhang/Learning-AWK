### gaussian distribution with various mean and standard deviation

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
```

```python
mu, sigma = -2, 0.45

x = np.linspace(-5, 5, 1000)
y1 = np.zeros(x.shape[0])
pdf = 1/(sigma * np.sqrt(2*np.pi)) * np.exp(-(x-mu)**2 / (2*sigma**2))
p = figure(plot_width=300, plot_height=300, y_range=(0, 1))
p.varea(x, y1=y1, y2=pdf, color="lawngreen")
show(p)
```
