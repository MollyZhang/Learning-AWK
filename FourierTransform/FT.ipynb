```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
t = np.arange(256, step=0.1)
y = np.cos(t/10) + np.sin(t/5)
```

```python
plt.plot(t, y)
plt.show()
```

```python
sp = np.fft.fft(y)
```

```python
freq = np.fft.fftfreq(y.shape[-1])
```

```python
plt.plot(freq, sp.real, label="real")
plt.plot(freq, sp.imag, label="imag")
plt.legend()
plt.show()
```
