```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
t = np.arange(10000, step=1)
y = np.cos(t/1) + np.sin(t/3)
```

```python
plt.figure(dpi=120)
plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y")
plt.show()
```

```python
sp = np.fft.fft(y)
```

```python
freq = np.fft.fftfreq(y.shape[-1])
```

```python
distance = (sp.real ** 2 + sp.imag ** 2) ** 0.5
```

```python
len(freq)
```

```python
r = 5000
# plt.plot(freq[:r], sp.real[:r], label="real")
# plt.plot(freq[:r], sp.imag[:r], label="imag")
plt.plot(freq[:r], distance[:r], label="distance")

plt.legend()
plt.show()
```
