```python
import tensorflow as tf
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
with tf.Session() as sess:
    x = sess.run(tf.truncated_normal((10000000,1),mean=0, stddev=0.1))
    x1 = sess.run(tf.random_normal((10000000,1), mean=0, stddev=0.1))
```

```python
ax = plt.hist(x, bins=100)
```

```python
ax = plt.hist(x1, bins=200)
```
