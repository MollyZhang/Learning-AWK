### Standard Dimension Reduction
in an n x m matrix where m < n

```python
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
from bokeh.io import output_notebook
from bokeh.plotting import figure, show

output_notebook()
```

```python
X1 = np.array([[1,2],
             [2,2.5],
             [3,3],
             [4,5]])
p = figure(plot_width=200, plot_height=200)
p.circle(X1[:,0], X1[:,1])
show(p)
```

```python
pca1 = PCA()
pca1.fit(X1)
pca1.transform(X1)
```

```python
pca1.explained_variance_ratio_
```
